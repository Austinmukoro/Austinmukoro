```python
company_data = company_data.rename(columns={'Rank': 'Ranking', 'Name': 'Company Name', 'Headquarters[note 1]':'Headquarters'})
```

```python
# Remove $ sign and replace with nothing from the Revenue and Profit column
```

```python
company_data.info()
```

```python
company_data['Revenue'] = company_data['Revenue'].str.replace('$','')
```

```python
company_data.head()
```

```python
company_data['Profit'] = company_data['Profit'].str.replace('$','')
```

```python
company_data.head()
```

```python
 # RENAME COLUMN NAME
```

```python
company_data.rename(columns={'Revenue':'Revenue($)','Profit':'Profit($)'},inplace=True)
```
